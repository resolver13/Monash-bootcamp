```python
# import libraries

import pandas as pd
import numpy as num
import matplotlib.pyplot as plt
df = pd.read_csv('data.csv')

print(df.head())     # for displaying first few rows of the dataframe

# calculate average, min, max
mean_temp = df["Temperature"].mean()
print(mean_temp)

min_temp = df["Humidity"].min()
max_temp = df["Humidity"].max()
print("min_temp: {} \n max_temp: {}".format(min_temp, max_temp))

std = df["Temperature"].std()
print(std)

# creating line plot graph
plt.figure(figsize=(8, 5))
plt.plot(df["Date"], df['Temperature'], marker="o", linestyle="-", color="b")
plt.title("Temperature data")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.xticks([1,10,20,30,40,50])
```
