```python
import matplotlib.pyplot as plt
```

```python
# for creating 50 random numbers
x = np.random.rand(50)
y = np.random.rand(50)

# gives more style in labeling
font1 = {'family':'serif','color':'green','size':20}
font2 = {'family':'times new roman','color':'blue','size':14}

# create a scatterplot
plt.scatter(x,y)

# set title and labels
plt.title("Random plot", fontdict = font1)
plt.xlabel("Age", fontdict = font2)
plt.ylabel("Percentage", fontdict = font2)

# overlay a line of best fit
m = np.polyfit(x, y, 1)
plt.plot(x, m[0] * x + m[1], color="red", label="Line of best fit")
plt.legend()

plt.show()
```
